```python
## BT biến, kiểu dữ liệu và toán tử
import turtle
t = turtle.Turtle()
turtle.bgcolor("lightblue")

# Bước 1: Vẽ nhà

t.penup()
t.bk(500)
t.pendown()

goc_vuong = 90
do_nghieng1 = 70.71


t.color("black", "blue")
t.begin_fill()
for i in range(4):
    t.fd(100)
    t.rt(goc_vuong)
t.end_fill()

t.color("black", "magenta")
t.begin_fill()
t.lt(goc_vuong/2)
t.fd(do_nghieng1)
t.rt(goc_vuong)
t.fd(do_nghieng1)
t.end_fill()

t.penup()
t.bk(do_nghieng1)
t.pendown()

t.color("black", "orange")
t.begin_fill()
t.left(goc_vuong*2/3)
t.fd(goc_vuong)
t.rt(goc_vuong*2/3)
t.fd(do_nghieng1)
t.rt(goc_vuong*4/3)
t.fd(goc_vuong)
t.end_fill()

t.lt(75)

t.penup()
t.fd(100)
t.pendown()

t.color("black", "yellow")
t.begin_fill()
t.lt(105)
t.fd(goc_vuong)
t.lt(75)
t.fd(100)
t.end_fill()

t.penup()
t.bk(100)
t.pendown()

t.lt(105)
t.fd(30)
t.rt(105)
t.end_fill()

t.penup()
t.fd(66)
t.pendown()

t.color("black", "brown")
t.begin_fill()
for i in range(2):
    t.lt(105)
    t.fd(33)
    t.lt(75)
    t.fd(33)
t.end_fill()

t.penup()
t.fd(33)
t.lt(105)
t.fd(66)
t.rt(15)
t.fd(66)
t.lt(goc_vuong)
t.fd(33)
t.pendown()

t.color("black", "palegreen")
t.begin_fill()
for i in range(2):
    t.fd(64)
    t.lt(goc_vuong)
    t.fd(33)
    t.lt(goc_vuong)
t.end_fill()
# chuyển sang vẽ mặt trời

t.lt(goc_vuong)
t.penup()
t.fd(200)
t.lt(goc_vuong)
t.fd(200)
t.pendown()


t.color("black", "yellow")
t.begin_fill()
for i in range(4):
    for i in range(1):
        t.lt(goc_vuong)
        t.fd(30)
        t.penup()
        t.bk(30)
        t.rt(goc_vuong)
        t.pendown()
    for i in range(45):
            t.rt(1)
            t.fd(0.5)
    t.lt(goc_vuong)
    t.fd(15)
    t.bk(15)
    
    t.rt(goc_vuong)
    for i in range(45):
        t.rt(1)
        t.fd(0.5)
t.end_fill()            
# chuyển sang vẽ cây
t.penup()
t.rt(goc_vuong)
t.fd(100)
t.rt(goc_vuong)
t.fd(100)
t.pendown()

t.color("black", "palegreen")
t.begin_fill()
for i in range(3):
    t.rt(goc_vuong/2)
    t.fd(goc_vuong*2/3)
    t.lt(135)
    t.fd(84.85)
    t.lt(135)
    t.fd(goc_vuong*2/3)
    t.lt(135)
    t.penup()
    t.fd(42.43)
    t.pendown()
t.end_fill()            
    
t.rt(goc_vuong)
t.penup()
t.fd(14.14)
t.pendown()

t.color("black", "brown")
t.begin_fill()
for i in range(2):
    t.lt(goc_vuong)
    t.fd(31)
    t.lt(goc_vuong)
    t.fd(21)
t.end_fill()     
```
